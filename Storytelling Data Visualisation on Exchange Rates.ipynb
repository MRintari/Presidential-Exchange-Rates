# Storytelling Data Visualisation on Exchange Rates
In this project, I will be predominantly focusing on the exchange rate between the euro and the American dollar.
## Explore Dataset
To begin analysis, we first need to import the dataset. This can be achieved using the pandas library, which can also be used for future analysis.

```python
import pandas as pd
exchange_rates = pd.read_csv('datasets/euro-daily-hist_1999_2022.csv')
```

We will now use the `head()` and `tail()` methods to inspect the first and last five rows, before using the `info()` method to gain a general overview of the dataset

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

Brief exploration of the dataset reveals that it contains 6311 rows and 41 columns, with many of these columns containing null values. We also see that the columns data types alter between being strings and floats.

## Data Cleaning
To make working with the dataset easier, the first step in cleaning will be renaming the `[US dollar ]` and `Period\Unit:` columns to `US_dollar` and `Time`, respectively.

```python
exchange_rates.rename(columns = {'[US dollar ]': 'US_dollar', 'Period\\Unit:':'Time'}, inplace = True)
exchange_rates.columns
```

I will now change the `Time` column to a `datetime` data type before sorting the values by `Time` in ascending order and resetting the index.

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace = True)
exchange_rates.reset_index(drop = True, inplace = True)
```

To complete analysis and visualisation using data from the `US_dollar` column, the data needs to be in numerical form. At the moment this is yet to be the case, evident from `.info()`'s output, with it currently being in string form.

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

Upon inspection of the different values present in the `US_dollar` column, we see 62 `-` values present. This is the reason for the column storing data in `object`/`string` form, therefore these values will be dropped.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Data Visualisation

```python
import matplotlib.pyplot as plt # Used for data visualisation
%matplotlib inline # Enables matplotlib visualisations in Jupyter Notebook

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar']) # Line graph
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.tail()
```

### Comparison of Euro-Dollar Rate under Presidents Bush, Obama and Trump
As I am to be comparing the euro-dollar rate under the presidencies of Bush, Obama, and Trump, data before the 20th January, 2001 is not needed, as this data would be prior to George Bush's inauguration. Following this same logic, we will not need any data from after Trump left office, on the 20th January 2021. Therefore, this data can be dropped. 

```python
bush_inauguration = '2001-01-20'
trump_departure = '2021-01-20'
euro_to_dollar = euro_to_dollar[euro_to_dollar['Time'] > pd.to_datetime(bush_inauguration)]
euro_to_dollar = euro_to_dollar[euro_to_dollar['Time'] < pd.to_datetime(trump_departure)]
euro_to_dollar.reset_index(drop = True, inplace = True)
```

```python
euro_to_dollar['US_dollar'].mean()
```
